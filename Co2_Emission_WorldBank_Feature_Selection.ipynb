- Initially the data is stored in a way that features for each year (2019-2023) correspond to different rows for a given country. We want to transform the data such that the featurees are along a column. This would be like transposing the DataFrame. We perform this by using a pivot table for year 2019. 
<br>
- Moreover, we see that all the features are of type 'object'. In order to perform numerical operations on these features we first convert them into to 'float' type. 

```python
df["2019 [YR2019]"]=df["2019 [YR2019]"].replace("..", None)
df["2019 [YR2019]"] = df["2019 [YR2019]"].astype(float)
```

```python
df_trans=pd.pivot_table(df, values="2019 [YR2019]", index="Country Name", columns="Series Code", aggfunc="mean", sort=False)
```

```python
df_trans=df_trans.iloc[:-48]
```

```python
df_trans.head()
```

```python
df_trans.info()
```

```python
df_trans.isnull().sum()
```

```python
drop_cols = []
for col in df_trans.columns:
    if df_trans[col].isnull().sum()>25:
        drop_cols.append(col)
```

```python
len(drop_cols)
```

```python
df_trans.drop(columns=drop_cols, axis=1, inplace=True)
df_trans.head()
```

```python
df_trans.info()
```

We see that there are many missing values in this data. To handle this we will first figure out the most representative set of features based on variance and correlation between features and explanability. Once we have narrowed down the set of features we want to build the LR model on, we can drop the missing values. 

## Feature variance

```python
from scipy.stats.mstats import trimmed_var
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import numpy as np
```

```python
trim_var = df_trans.apply(trimmed_var, limits=(0.1,0.1)).sort_values()
```

```python
trim_var.astype("float").tail()
```

```python
trim_var.astype("float").head()
```

```python
trim_var.tail().astype("float").plot.bar()
```

## Standardize feature scale

```python
# Instantiate transformer
ss = StandardScaler()

# Transform `X`
X_scaled_data = ss.fit_transform(df_trans)

# Put `X_scaled_data` into DataFrame
X_scaled = pd.DataFrame(X_scaled_data, columns=df_trans.columns, index=df_trans.index)

print("X_scaled type:", type(X_scaled))
print("X_scaled shape:", X_scaled.shape)
X_scaled.head()
```

```python
X_scaled_summary = X_scaled.aggregate(["mean","std"]).astype(int)

print("X_scaled_summary type:", type(X_scaled_summary))
print("X_scaled_summary shape:", X_scaled_summary.shape)
X_scaled_summary
```

## Correlation anlaysis

```python
# Create correlation matrix
correlation = X_scaled.select_dtypes("number").corr().abs()
correlation
# Plot heatmap of `correlation`
sns.heatmap(correlation);
```

```python
# Drop features with high correlation
threshold = 0.55
# Select upper triangle of correlation matrix
upper = correlation.where(np.triu(np.ones(correlation.shape), k=1).astype(bool))

# Find features with correlation greater than threshold
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
```

```python
X_reduced = X_scaled.drop(to_drop, axis=1)
```
